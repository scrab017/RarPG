```python
import numpy as np
import pandas as pd
import scanpy as sc
from scipy import sparse
```

```python
sc.settings.verbosity = 3
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor='white')
results_file = '../results/pbmc68k_op.h5ad'
```

```python
sc.logging.print_memory_usage()
```

```python
%%time
path = '../data/processed/pbmc68k/'
adata = sc.read(path + 'matrix.mtx', cache=True).T
adata.var_names = pd.read_csv(path + 'genes.csv')['0']
adata.obs_names = pd.read_csv(path + 'barcodes.csv')['0']
```

```python
adata
```

```python
adata.var_names_make_unique()
```

```python
adata.obs['bulk_labels'] = pd.read_csv(path + 'celltype.csv')['celltype'].values
```

### HVGS

```python
%%time
sc.pp.highly_variable_genes(adata, flavor='seurat_v3', n_top_genes=1000)
```

```python
%%time
sc.pl.highly_variable_genes(adata)
```

```python
%%time
sc.pl.highly_variable_genes(adata, log=True)
```

```python
adata
```

```python
%%time
adata = adata[:, adata.var.highly_variable]
sc.pp.normalize_per_cell(adata) 
sc.pp.log1p(adata, base=2)
sc.pp.scale(adata)
```

```python
adata
```

### Compute Neighbours

```python
%%time
sc.pp.neighbors(adata, metric='euclidean', n_neighbors=10, random_state=1)
sc.logging.print_memory_usage()
```

### UMAP

```python
%%time
sc.tl.umap(adata, min_dist=0.5, spread=1.0, random_state=1, n_components=2)
```

```python
%%time
sc.pl.umap(adata, color='bulk_labels')
```

### PaCMAP

```python
import pacmap
```

```python
embedding = pacmap.PaCMAP(n_dims=2, n_neighbors=10, MN_ratio=0.3, FP_ratio=3.3) 
```

```python
%%time
X_transformed = embedding.fit_transform(adata.X, init="pca")
```

```python
import matplotlib.pyplot as plt
```

```python
adata.obs['bulk_labels'].cat.codes.values
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 12))

scatter = ax.scatter(X_transformed[:, 0], X_transformed[:, 1], cmap="Spectral", c=adata.obs['bulk_labels'].cat.codes.values,s=0.6
          ,label=adata.obs['bulk_labels'].values)

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="best", title="cell types")
ax.add_artist(legend1)
plt.savefig('figures/PaCMAP_pbmc.jpg',bbox_inches='tight',dpi=600)
```
